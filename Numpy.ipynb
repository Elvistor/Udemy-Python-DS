```python
# Comparativa numpy - python list
from datetime import datetime
py_list = [i for i in range(1000000)]
start = datetime.now()
py_list = [i+3 for i in py_list]
end = datetime.now()
print(end - start)
```

```python
import sys
s = range(1000)
print(s)
print("Tamaño de cada elemento de la lista en bytes: ", sys.getsizeof(s))
print("Tamaño total de la lista en bytes: ", sys.getsizeof(s)*len(s))
```

```python
nS = np.arange(1000)
print("Tamaño de cada elemento de la lista en bytes: ", nS.itemsize)
print("Tamaño total de la lista en bytes: ", nS.itemsize*nS.size)
```
