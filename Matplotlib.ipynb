```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.__version__
```

```python
edad_x = [20,25,30,35,40,45,50,55,60,65]
salario_cd_y = [20000, 30000, 33000, 40000, 42000, 45000, 47000, 60000, 70000, 80000]
salario_fed_y = [10000, 15000, 17000, 27000, 30000, 32000, 35000, 40000, 49000, 50000]
salario_bed_y = [15000, 20000, 25000, 27000, 35000, 40000, 45000, 50000, 55000, 60000]
```

```python
plt.plot(edad_x, salario_cd_y)
plt.title('Salario Medio en U$D de un DS por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')
plt.show()
```

```python
plt.plot(edad_x, salario_cd_y, label = 'Científico de Datos')
plt.plot(edad_x, salario_fed_y, label = 'Front End Desarrollador')

plt.title('Salario Medio en U$D por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')

plt.legend()

plt.show()
```

```python
plt.plot(edad_x, salario_cd_y, label = 'Científico de Datos', linewidth=3, marker='.', markersize=12, linestyle = '--', color='#d62728')
plt.plot(edad_x, salario_fed_y, label = 'Front End Desarrollador')
plt.plot(edad_x, salario_bed_y, label = 'Back End Desarrollador')

plt.title('Salario Medio en U$D por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')

plt.legend()

plt.grid(True)

plt.show()
```

```python
import numpy as np

plt.plot(edad_x, salario_cd_y, label = 'Científico de Datos', linewidth=3, marker='.', markersize=20, color='yellow')
plt.plot(edad_x, salario_fed_y, label = 'Front End Desarrollador')
plt.plot(edad_x, salario_bed_y, label = 'Back End Desarrollador')

salario_medio = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis=0)/3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle = '--', color='white', marker='.', markersize=30)

plt.title('Salario Medio en U$D por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')

plt.legend()

plt.grid(False)

plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label = 'Back End Desarrollador')

salario_medio = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis=0)/3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle = '--', color='black', marker='.', markersize=10)

plt.fill_between(edad_x, salario_bed_y, salario_medio, interpolate = True, alpha = 0.33, label = 'Between')

plt.title('Salario Medio en U$D por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')

plt.legend()

plt.grid(False)

plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label = 'Back End Desarrollador')

salario_medio = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis=0)/3
plt.plot(edad_x, salario_medio, label = 'Salario Medio', linestyle = '--', color='black', marker='.', markersize=10)

plt.fill_between(edad_x, salario_bed_y, salario_medio, where=(salario_bed_y > salario_medio),
                 interpolate = True, alpha = 0.33, label = 'Between')

plt.fill_between(edad_x, salario_bed_y, salario_medio, where=(salario_bed_y <= salario_medio),
                 interpolate = True, color='yellow', alpha = 0.33, label = 'Between')

plt.title('Salario Medio en U$D por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')

plt.legend()

plt.grid(False)

plt.show()
```

```python
width = 0.1

edad_np_x = np.arange(len(edad_x))

plt.bar(edad_np_x -2*width, salario_cd_y, label = 'Científico de Datos', color='red', width = width)
plt.bar(edad_np_x, salario_fed_y, label = 'Front End Desarrollador', width = width)
plt.bar(edad_np_x +2*width, salario_bed_y, label = 'Back End Desarrollador', width = width)

salario_medio = np.sum([salario_cd_y, salario_fed_y, salario_bed_y], axis=0)/3
plt.bar(edad_np_x -4*width, salario_medio, label = 'Salario Medio', color='black', width = 2*width)

plt.title('Salario Medio en U$D por Edad')
plt.xlabel('Edad')
plt.ylabel('Salario medio U$D')

plt.xticks(ticks = edad_np_x, labels = edad_x)

plt.legend()

plt.grid(False)

plt.show()
```
